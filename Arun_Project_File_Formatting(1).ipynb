```python
data = data[["Columns","Data"]]
print(data.head(35))
```

```python
l = ['nan','----','Accept','Accept-Charset','Accept-Encoding','Accept-Language','Authorization','Cache-Control','Client-ip','Connection','Cookie','Cookie2','Date','ETag','End - Id','Expect','From','GET','Host','If-Match','If-Modified-Since','If-None-Match','If-Range','If-Unmodified-Since','MIME-Version','Max-Forwards','Pragma','Proxy-Authorization','Range','Referer','Start - Id','TE','Trailer','Transfer-Encoding','UA-CPU','UA-Color','UA-Disp','UA-OS','UA-Pixels','Upgrade','User-Agent','Via','Warning','X-Forwarded-For','X-Serial-Number','class','~~~~~','Content-Encoding','Content-Language','Content-Length','Content-Location','Content-MD5','Content-Type','Expires','Last-Modified','PUT','PUT Body','POST','POST Body']
for i,rows in data.iterrows():
    if re.findall("\AGET", str(rows['Columns'])) or re.findall("\APOST", str(rows['Columns'])) or re.findall("\APUT", str(rows['Columns'])):
        #print("i")
        X = rows['Columns'].split()
       # print(X)
       # print(rows['Columns'])
        rows['Columns'] = X[0]
        rows['Data'] = X[1]
        
    if str(rows['Columns']) not in l and str(rows['Data']) != '0':
        rows['Columns'] = str(rows['Columns']) +":"+ str(rows['Data'])
        rows['Data'] = str(0)
        print(rows['Columns'])

        
    


        
```

```python
new_row ={}
new_data = pd.DataFrame()
for i,rows in data.iterrows():
    new_row[rows['Columns']] = rows['Data']
    if rows['Columns'] == 'End - Id':
        new_row[rows['Columns']] = rows['Data']
        string = ""
        if 'GET' in new_row.copy().keys():
            for key,value in new_row.copy().items():
                if value == '0':
                    del new_row[key]
                 
        if 'POST' in new_row.copy().keys():
            for key,value in new_row.copy().items():
                if value == '0':
                    string += str(key)
                    del new_row[key]
            new_row['POST Body'] = string
            
        if 'PUT' in new_row.copy().keys():
            for key,value in new_row.copy().items():
                if value == '0':
                    string += str(key)
                    del new_row[key]
            new_row['PUT Body'] = string
        print(i)
        new_data = new_data.append(new_row, ignore_index = True)
        new_row = {}
        continue
    
```

```python
#print(new_data.head(2))
new_data.to_csv("xml_test_formatted.csv")
```
